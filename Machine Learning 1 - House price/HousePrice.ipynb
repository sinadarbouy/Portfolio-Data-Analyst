```python
import pandas as pd
```

```python
#Data Description

file_path = 'data/melb_data.csv'
data = pd.read_csv(file_path)

data.describe()

#Maximum Year = 2018
#AVG Size = 3991
```

```python
data.columns

#Fix Data
data= data.dropna(axis=0)
```

```python
#Prediction Target
y = data.Price

#Select Features
data_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']

x= data[data_features]

x.head()
```

```python
#model

from sklearn.tree import DecisionTreeRegressor

#define model
model = DecisionTreeRegressor(random_state=1)

#fit model
model.fit(x,y)
```

```python
#prediction
model.predict(x.head())
```
